```python
# Carga el dataset de trabajo
import pandas as pd
import numpy as np
df1 = pd.read_csv('Lab2.csv')
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
df1.head(1)
condicion1 = df1['release_year']== 2021
condicion2 = df1['type']== 'Movie'
df2 = df1[condicion1 & condicion2]
df2.head(1)
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
df1.head(1)
condicion1 = df1['duration']== '90 min'
df3 = df1[condicion1].filter(items=['title','duration'])
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
df1.head(1)
condicion1 = df1['country']== 'India'
condicion2 = df2['release_year']>2018
df4 = df1[condicion1 & condicion2].filter(items=['title'])
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
df1.head(1)
```

```python

condicion1 = df1['rating']== 'TV-14'
condicion2 = df2['rating'] == 'R'
df5 = df1[condicion1 | condicion2].filter(items=['title','rating'])
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
df1.head(1)
```

```python

condicion1 = df1['director']== 'Christopher Nolan'
condicion2 = df2['director'] == 'Martin Scorsese'
df6 = df1[condicion1 | condicion2].filter(items=['title','director']).sum()
df6.value_counts().reset_index()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
df1.head(1)
```
